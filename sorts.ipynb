```python
# 第一步对相同长度的两个顺序数列归并排序
# 顺序的假定很重要 
# 假设和实现的分离
def merge(x,y):
    len_x = len(x)
    len_y = len(y)
    i,j=0,0
    c=[]
#     print('c:',c)
    while (i<len_x) & (j<len_y):
        if x[i]<y[j]:
            c.append(x[i])
            i=i+1
#             print('cx:',c)
        else:
            c.append(y[j])
            j=j+1
#             print('cy',c)
    if i>j:
        c.extend(y[j:len_y])
    else:
        c.extend(x[i:len_x])
#     print('c:',c)
    return c
# 方案一不考虑形参实参共变的情况：
def merge_divide(a):
    len_a=len(a)
    up=int(len_a/2)
    a1=a[:up].copy()
    a2=a[up:].copy()
    if up==1:
        return merge(a1,a2)
    else:
        newa1=merge_divide(a1)
        newa2=merge_divide(a2)
        return merge(newa1,newa2)
        
```

## 快速排序算法

```python
# 方案一是带返回的
# 函数定义区
# 第一步实现大小交换
def swap(x,i,j):
    x[i],x[j]=x[j],x[i]
# 第二步实现对首位的一次移动
def fast_sort(x):
    len_x = len(x)
    i=0
    j=len_x-1
    k=x[0]
    while i!=j:
        while(i<j and k<x[j]):
            j=j-1
        swap(x,i,j)
        while(i<j and k>x[i]):
            i=i+1
        swap(x,i,j)
        x1=x[0:j]
        x2=x[j+1:len_x]
    if len(x1)>1:
            x[0:j]=fast_sort(x1)
    if len(x2)>1:
            x[j+1:len_x]=fast_sort(x2)
    return x
```

## 求前m大的数从大到小输出

```python
# 交换列表元素
def swap(x,i,j):
    x[i],x[j]=x[j],x[i]
# 第一步利用阈值分离目标区域
# 方案二 用位置方法
def arrangeRight(x,m,s,e,yuan):
    a=x[s:e+1]
    print('before',x)
    n=0
    len_x = len(x)
    len_a = len(a)
    i=0
    j=len_a-1
    k=a[0]
    while i!=j:
        n=n+1
        if n>20:
            sys.exit()
        while(i<j and k<a[j]):
            j=j-1
        swap(a,i,j)
        while(i<j and k>a[i]):
            i=i+1
        swap(a,i,j)
    
    x[s:e+1]=a
    weizhi=j+s
    bge=j+s
    age=len_x-bge
    
    if age>m:
        print('age',age)
        print('右边')
        arrangeRight(x,yuan,j+s+1,len_x,yuan)
    elif age<m:
        print('左边')
        arrangeRight(x,m-age,0,bge-1,yuan)
    else:
        pass
    print('weizhi,after',weizhi,x)
```
